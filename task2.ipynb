```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/BRCA.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#find the missing values
df.isna().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df['Protein1'].fillna(method='ffill',inplace=True)
```

```python
df['Protein2'].fillna(df['Protein2'].median(),inplace=True)
```

```python
df['Protein3'].fillna(df['Protein3'].median(),inplace=True)
```

```python
df=df.fillna(method='bfill')
```

```python
df['Protein4'].fillna(df['Protein4'].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
from statsmodels.tsa.arima.model import ARIMA
model=ARIMA(df['Protein1'],order=(1,1,1))
model_fit=model.fit()
print(model_fit.summary)
model_fit.plot_diagnostics(figsize=(7,6))
plt.show()
forecast=model_fit.forecast(steps=10)
forecast
```

```python
from statsmodels.tsa.arima.model import ARIMA
model=ARIMA(df['Protein4'],order=(1,1,1))
model_fit=model.fit()
print(model_fit.summary)
model_fit.plot_diagnostics(figsize=(7,6))
plt.show()
forecast=model_fit.forecast(steps=10)
forecast
```
